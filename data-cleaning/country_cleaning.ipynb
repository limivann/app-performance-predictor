```python
import pandas as pd
import numpy as np
```

```python
developer_df = pd.read_csv("./input/dev_addr.csv")
developer_df
```

```python
developer_df.dropna(inplace = True)
developer_df = developer_df[developer_df["DEVELOPER_ADDRESS_PRED"] != "undefined"]
developer_df["COUNTRY"] = ""
developer_df.shape
```

```python
for i in range(len(developer_df)):
    address = developer_df["DEVELOPER_ADDRESS_PRED"].iloc[i]
    front = address.split(",")[0]
    front = front.upper()
    countryList = ['JAPAN', 'KOREA', 'SOUTH KOREA', 'TAIWAN', 'HONG KONG', 'CHINA', 'CN', 'JP', 'KR', 'TW']
    if front in countryList:
        developer_df["COUNTRY"].iloc[i] = front
        
    else:
        last = address.split(", ")[-1]
        last = last.upper()
        if last.isdigit():
            last = address.split(", ")[-2]
            last = last.upper()
            
        last = ''.join([i for i in last if not i.isdigit()]) #remove digits from last substring (after comma)
        if "香港" in address:
            last = "HONG KONG"
        if "中国" in address:
            last = "CHINA"
        if "UK" in address:
            last = "UNITED KINGDOM"
        if "US" in address:
            last = "USA"
        developer_df["COUNTRY"].iloc[i] = last
```

```python
#Clear emirates
for i in range(len(developer_df)):
    address = developer_df["DEVELOPER_ADDRESS_PRED"].iloc[i]
    last = address.split("- ")[-1]
    last = last.upper()
    if "EMIRATES" in last:
        developer_df["COUNTRY"].iloc[i] = "UNITED ARAB EMIRATES" 
    if "UAE" in last:
        developer_df["COUNTRY"].iloc[i] = "UNITED ARAB EMIRATES" 
```

```python
#Clear Hong Kong and Japan
for i in range(len(developer_df)):
    address = developer_df["DEVELOPER_ADDRESS_PRED"].iloc[i]
    address = address.upper()
    if "JAPAN" in address:
        developer_df["COUNTRY"].iloc[i] = "JAPAN" 
    if "HONG KONG" in address:
        developer_df["COUNTRY"].iloc[i] = "HONG KONG"
```
